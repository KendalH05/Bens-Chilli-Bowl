```python
import pandas as pd
import numpy as np 
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')



```

```python
Housing_prices = pd.read_csv("zillow_prices.csv")
print(Housing_prices.shape)
```

```python
Housing_prices.sample(n=5)
```

```python
Housing_prices

dc_Housing = Housing_prices[Housing_prices.StateName == 'DC']
dc_Housing
#md_unemployment = md_unemployment.filter(items = ['Year','Month', "State", 'County', 'Rate'])
```

```python
dc_Housing_edited = dc_Housing.drop(['RegionID', 'SizeRank', 'RegionType'], axis=1)
dc_Housing_edited
```

```python
dc_Housing_edited = dc_Housing_edited.melt(id_vars=["RegionName", "StateName"], 
        var_name="Date", value_name="Value")
```

```python
list(dc_Housing_edited.columns)
#dc_Housing_edited['Date'] = dc_Housing_edited['Date'].str.replace('/','-')
```

```python

dc_Housing_edited['Year'] = pd.DatetimeIndex(dc_Housing_edited['Date']).year
Date = dc_Housing_edited['Year'].to_numpy()
Value = dc_Housing_edited['Value'].to_numpy()
plt.plot(Date,Value )
plt.title("DC Housing prices")
plt.xlabel('Date')
plt.ylabel("Price")
plt.show()
```
